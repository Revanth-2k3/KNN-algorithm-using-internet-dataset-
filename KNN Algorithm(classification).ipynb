```python
# Drop rows with missing values in critical columns
cleaned_data = data.dropna(subset=["Rate (WB)", "Rate (ITU)", "Users (CIA)"])
```

```python
# Define features and target
features = cleaned_data[["Rate (WB)", "Rate (ITU)", "Users (CIA)"]]
# Classify as 'High' or 'Low' internet usage based on Rate (WB)
threshold = 50  # Example threshold
cleaned_data["Internet_Usage_Class"] = cleaned_data["Rate (WB)"].apply(
    lambda x: "High" if x >= threshold else "Low"
)
target = cleaned_data["Internet_Usage_Class"]
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create and train the KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)  # Using 5 neighbors as default
knn.fit(X_train_scaled, y_train)
```

```python
# Predict on the test set
y_pred = knn.predict(X_test_scaled)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
# Print results
print(f"Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:\n")
print(report)
```
